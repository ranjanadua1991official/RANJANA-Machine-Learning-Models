```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/data.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df.head()
```

```python
df.columns
```

```python
dataset=df.drop(columns='Unnamed: 32')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
en=LabelEncoder()
```

```python
dataset['diagnosis']=en.fit_transform(dataset['diagnosis'])
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset['id'].value_counts()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
dataset.isnull().sum()
```

```python
x=dataset.drop(columns='diagnosis', axis=1)
y=dataset['diagnosis']
```

```python
#Apply train_test split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
sc=StandardScaler()
```

```python
x_train_scale=sc.fit_transform(x_train)
x_test_scale=sc.transform(x_test)
```

```python
model=LogisticRegression()
```

```python
model
```

```python
model.fit(x_train_scale,y_train)
```

```python
y_train_pred=model.predict(x_train_scale)
y_test_pred=model.predict(x_test_scale)
```

```python
Train_accuracy=accuracy_score(y_train,y_train_pred)
Test_accuracy=accuracy_score(y_test,y_test_pred)
```

```python
print(f'The Train accuracy of model is {Train_accuracy}')
print(f'The Test accuracy of model is {Test_accuracy}')
```
