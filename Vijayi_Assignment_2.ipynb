#Model Fine Tuning

```python
from sentence_transformers import SentenceTransformer, InputExample, losses
model = SentenceTransformer("all-MiniLM-L6-v2")
# Create training examples (query, positive quote)
train_examples = [InputExample(texts=["quote about hope", "Hope is the thing with feathers..."], label=1.0)]
trainer = losses.MultipleNegativesRankingLoss(model)
model.fit(train_examples, epochs=3)
model.save("fine_tuned_quote_model")
```

#Build RAG Pipeline

```python
from sentence_transformers import util
embeddings = model.encode(dataset["quote"].tolist())
faiss_index = util.faiss_index(embeddings)  # Build index
```

```python
query = "quotes about courage by women authors"
query_embedding = model.encode(query)
scores, indices = faiss_index.search(query_embedding, k=3)
retrieved_quotes = df.iloc[indices[0]]
# Feed retrieved quotes + query to LLM for summary/response
```
