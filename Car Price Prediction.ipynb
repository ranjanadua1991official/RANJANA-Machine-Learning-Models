```python
#Encoding the categorical data into numerical data
```

```python
car_dataset.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_dataset.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
```

```python
car_dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
```

```python
car_dataset.describe()
```

```python
car_finaldataset=car_dataset.drop(columns='Car_Name')
```

```python
car_finaldataset
```

```python
#Separate variables and target
x=car_finaldataset.drop(columns='Selling_Price')
y=car_finaldataset['Selling_Price']
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Perform train_test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

# Model Selection

```python
model=LinearRegression()
```

```python
print(model)
```

```python
model.fit(x_train_sc,y_train)
```

```python
y_train_pred=model.predict(x_train_sc)
y_test_pred=model.predict(x_test_sc)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
#Find R2 score of train and test data
r2_train=r2_score(y_train_pred,y_train)
r2_test=r2_score(y_test_pred,y_test)
```

```python
print(f'The R2 Score of training data is {r2_train}')
print(f'The R2 Score of testing data is {r2_test}')
```

```python
mse_train=mean_squared_error(y_train_pred,y_train)
mse_test=mean_squared_error(y_test_pred,y_test)
```

```python
print(f'The MSME Score of training data is {mse_train}')
print(f'The MSME Score of testing data is {mse_test}')
```
